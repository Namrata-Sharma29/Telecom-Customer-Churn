```python

# Random Forest Regressor for Churn Prediction

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 2. Encode target variable 'Churn' (Yes=1, No=0)
df['ChurnFlag'] = df['Churn'].apply(lambda x: 1 if str(x).strip().lower() == 'yes' else 0)

# 3. Keep only the 3 features requested
features = ['Contract', 'Payment Method', 'Internet Type']
df_model = df[features + ['ChurnFlag']].copy()

# 4. Encode categorical features
le = LabelEncoder()
for col in features:
    df_model[col] = le.fit_transform(df_model[col].astype(str))

# 5. Features (X) and Target (y)
X = df_model[features]
y = df_model['ChurnFlag']

# 6. Train-Test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# 7. Feature Scaling (optional but helps normalization)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 8. Initialize and train RandomForestRegressor
rf_reg = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    max_depth=None,
    min_samples_split=2
)
rf_reg.fit(X_train_scaled, y_train)

# 9. Predictions
y_pred = rf_reg.predict(X_test_scaled)

# 10. Evaluate
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"RÂ² Score: {r2:.4f}")

# 11. Show first few predictions side-by-side
pred_df = pd.DataFrame({
    'Actual Churn': y_test.values,
    'Predicted (Score)': y_pred
})
print("\nSample Predictions:\n", pred_df.head(10))
```
